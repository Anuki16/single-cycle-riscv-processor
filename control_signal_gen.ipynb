```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("riscv_controller.csv", dtype=str)
df.dropna(how="all", inplace=True)
```

```python
df
```

```python
for i in range(len(df)):
    row = df.iloc[i]
    if type(row.Opcode) == str:
        addr = row.Opcode[:-2]
        addr += row.func3 if row.func3 != '-' else '000'
        addr += row.func7[1] if row.func7 != '-' else '0'
    else:
        addr = row.micro_addr

    assert addr == row.micro_addr
    
    print(f"\tassign control_store[9'b{addr}] = {{1'b{row.write_en},	4'b{row.alu_sel},	1'b{row.alu_b_sel},	1'b{row.alu_a_sel},	1'b{row.mem_write},	1'b{row.mem_read},	2'b{row.load_store_type},	1'b{row.load_unsigned},	2'b{row.write_src_sel},	3'b{row.branch_type},\t1'b{row.stay},\t1'b{row.memcpy_store},\t1'b{row.counter_en},\t1'b{row.counter_sel},\t9'b{row.next_uinst}}}; // {row.op}")
```
